# drive에서 local로 복사

```python
import os
import time

# 1. Google Drive → Colab 복사
!cp "/content/drive/MyDrive/Colab Notebooks/APID/iu_xray_r2gen.zip" /content/

# 2. 압축 해제 + 시간 측정
start = time.time()
!unzip -q /content/iu_xray_r2gen.zip -d /content
print("압축 해제 완료:", round(time.time() - start, 2), "초")

# 3. 폴더별 이미지 수 출력
base_path = '/content/iu_xray/images'

total_images = 0
for folder in sorted(os.listdir(base_path)):
    folder_path = os.path.join(base_path, folder)
    if os.path.isdir(folder_path):
        num_images = len([
            f for f in os.listdir(folder_path)
            if f.lower().endswith(('.jpg', '.jpeg', '.png'))
        ])
        print(f"{folder}: {num_images}장")
        total_images += num_images

print(f"\n총 이미지 수: {total_images}장")
```
